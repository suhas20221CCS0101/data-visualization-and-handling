```python
import pandas as pd
import numpy as np
path="/content/drive/MyDrive/Flickerimages/BL-Flickr-Images-Book.csv"
df=pd.read_csv(path)
print(df.head(5))
```

```python
to_drop=['Edition Statement' ,'Corporate Author','Contributors','Corporate Author','Former owner','Engraver','Issuance type','Shelfmarks']
df.drop(to_drop, inplace=True, axis=1)
df.head()
```

```python
df.set_index('Identifier', inplace = True)
df.head()
df['Date of Publication'].head(25)
```

```python
unwanted_characters=['[',',','-']
def clean_dates(item):
  dop=str(item.loc['Date of Publication'])
  if dop == 'nan' or dop[0]=='[':
      return np.NaN
  for characters in unwanted_characters:
    if characters in dop:
           character_index=dop.find(characters)
           dop=dop[:character_index]
    return dop
df['Date of publication'] = df.apply(clean_dates, axis=1)
df.head()
df['Date of publication'].head(25)
```
